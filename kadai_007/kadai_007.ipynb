```python
# 商品データ
data = {
    '注文番号': [101, 102, 103, 104, 105, 106],
    '商品番号': [1, 2, 1, 3, 2, 1],
    '注文数': [2, 1, 3, 1, 2, 1]
}
df = pd.DataFrame(data)

# カテゴリデータ
category_data = {
    '商品番号': [1, 2, 3],
    'カテゴリー': ['弁当', 'デザート', '飲料']
}
category_df = pd.DataFrame(category_data)

# マージ
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
```

```python
df['カテゴリー'].value_counts().plot(kind='bar')
plt.title("カテゴリーの出現頻度")
plt.xlabel("カテゴリー")
plt.ylabel("件数")
plt.show()
```
