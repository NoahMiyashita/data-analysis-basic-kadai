```python
# Step1: matplotlibとnumpyをインポート
import matplotlib.pyplot as plt
import numpy as np

# Step2: 描画キャンバスを2行3列で用意（計6領域）
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6), tight_layout=True)

# Step3-1: 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('Line Plot')

# Step3-2: sin関数
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title('Sine Wave')

# Step3-3: ヒストグラム
data2 = np.random.randn(1000)
axes[0, 2].hist(data2, bins=30)
axes[0, 2].set_title('Histogram')

# Step3-4: 散布図
x3 = np.random.rand(50)
y3 = np.random.rand(50)
axes[1, 0].scatter(x3, y3)
axes[1, 0].set_title('Scatter Plot')

# 他の2領域（[1,1]と[1,2]）は空でOK

# Step4: グラフを表示
plt.show()
```
